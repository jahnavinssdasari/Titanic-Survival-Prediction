```python
#Adding Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.info()
```

```python
df.drop(columns=['Cabin','Ticket','Name','PassengerId','Embarked'],inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Age']=df['Age'].astype(int)
```

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
plt.title('Survival of people')
plt.xlabel("No of people Survived")
plt.ylabel("Count")
plt.show()
```

```python
sns.barplot(x=df['Survived'],y=df['Fare'])
plt.show()
```

```python
a=df['Survived'].value_counts().index.tolist()
b=df['Survived'].value_counts().values.tolist()
print(a,b)
```

```python
plt.pie(b,labels=a,autopct='%.1f%%')
plt.title('Survival Percentage on Board')
plt.legend()
plt.show()
```
